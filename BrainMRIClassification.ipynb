```python
modelEffn_0 = efn.EfficientNetB0(weights='imagenet',include_top=True)
modelEffn_1 = efn.EfficientNetB1(weights='imagenet',include_top=True)
modelEffn_2 = efn.EfficientNetB2(weights='imagenet',include_top=True)
modelEffn_3 = efn.EfficientNetB3(weights='imagenet',include_top=True)
modelEffn_4 = efn.EfficientNetB4(weights='imagenet',include_top=True)

featureCountLists = [500,1000,1500,2000,2500,3000,3500,4000,4500]
BATCH_SIZE  = 32


for fold in range(1,6):
  dfData = pd.read_csv("labels/labels_fold_{}.csv".format(fold))

  dfTrain = dfData[dfData['imageType']=="train"] 
  dfTest = dfData[dfData['imageType']=="test"] 
  len(dfTrain)

  datagen=ImageDataGenerator(rescale=1./255)

  train_generator=datagen.flow_from_dataframe(dataframe=dfTrain,directory="/content", x_col="id",y_col="label",
  batch_size=BATCH_SIZE,shuffle=False,class_mode='categorical',target_size=(224,224),classes=['no', 'yes'])

  test_generator=datagen.flow_from_dataframe(dataframe=dfTest,directory="/content",x_col="id",y_col="label",batch_size=BATCH_SIZE,
  shuffle=False,class_mode='categorical',target_size=(224,224), classes=['no', 'yes'])

  
  image_count = len(dfTrain)

  x_train_Effn_0 = modelEffn_0.predict_generator(train_generator, verbose=1)
  x_test_Effn_0 = modelEffn_0.predict_generator(test_generator, verbose=1)  

  train_generator=datagen.flow_from_dataframe(dataframe=dfTrain,directory="/content", x_col="id",y_col="label",
  batch_size=BATCH_SIZE,shuffle=False,class_mode='categorical',target_size=(240,240),classes=['no', 'yes'])

  test_generator=datagen.flow_from_dataframe(dataframe=dfTest,directory="/content",x_col="id",y_col="label",batch_size=BATCH_SIZE,
  shuffle=False,class_mode='categorical',target_size=(240,240), classes=['no', 'yes'])

  x_train_Effn_1 = modelEffn_1.predict_generator(train_generator, verbose=1)
  x_test_Effn_1 = modelEffn_1.predict_generator(test_generator, verbose=1)  


  train_generator=datagen.flow_from_dataframe(dataframe=dfTrain,directory="/content", x_col="id",y_col="label",
  batch_size=BATCH_SIZE,shuffle=False,class_mode='categorical',target_size=(260,260),classes=['no', 'yes'])

  test_generator=datagen.flow_from_dataframe(dataframe=dfTest,directory="/content",x_col="id",y_col="label",batch_size=BATCH_SIZE,
  shuffle=False,class_mode='categorical',target_size=(260,260), classes=['no', 'yes'])

  x_train_Effn_2 = modelEffn_2.predict_generator(train_generator, verbose=1)
  x_test_Effn_2 = modelEffn_2.predict_generator(test_generator, verbose=1)  
  
  train_generator=datagen.flow_from_dataframe(dataframe=dfTrain,directory="/content", x_col="id",y_col="label",
  batch_size=BATCH_SIZE,shuffle=False,class_mode='categorical',target_size=(300,300),classes=['no', 'yes'])

  test_generator=datagen.flow_from_dataframe(dataframe=dfTest,directory="/content",x_col="id",y_col="label",batch_size=BATCH_SIZE,
  shuffle=False,class_mode='categorical',target_size=(300,300), classes=['no', 'yes'])

  x_train_Effn_3 = modelEffn_3.predict_generator(train_generator, verbose=1)
  x_test_Effn_3 = modelEffn_3.predict_generator(test_generator, verbose=1)  

  train_generator=datagen.flow_from_dataframe(dataframe=dfTrain,directory="/content", x_col="id",y_col="label",
  batch_size=BATCH_SIZE,shuffle=False,class_mode='categorical',target_size=(380,380),classes=['no', 'yes'])

  test_generator=datagen.flow_from_dataframe(dataframe=dfTest,directory="/content",x_col="id",y_col="label",batch_size=BATCH_SIZE,
  shuffle=False,class_mode='categorical',target_size=(380,380), classes=['no', 'yes'])  

  x_test_Effn_4 = modelEffn_4.predict_generator(test_generator, verbose=1)  
  x_train_Effn_4 = modelEffn_4.predict_generator(train_generator, verbose=1)

  x_train = np.concatenate((x_train_Effn_0,x_train_Effn_1,x_train_Effn_2,x_train_Effn_3,x_train_Effn_4),axis=1)
  x_test = np.concatenate((x_test_Effn_0,x_test_Effn_1,x_test_Effn_2,x_test_Effn_3,x_test_Effn_4),axis=1)

  y_train = train_generator.classes
  y_test  = test_generator.classes

  scaler = StandardScaler()
  scaler.fit(x_train)
  x_train = scaler.transform(x_train)
  x_test = scaler.transform(x_test)

  classifier =  SVC(kernel="linear")
  classifier.fit(x_train, y_train)
  y_head = classifier.predict(x_test)
  score = accuracy_score(y_test, y_head)
  print("Fold No : {} - Feature Count :{} - Accuracy {} ".format(fold,5000,score))

  for featureCount in featureCountLists:
    KBest = SelectKBest(f_classif, k=featureCount)
    KBest.fit_transform(x_train, y_train)
    selectedFuture = KBest.get_support(indices=True)

    x_train_selected = x_train[:,selectedFuture]
    x_test_selected = x_test[:,selectedFuture]  

    classifier =  SVC(kernel="linear")
    classifier.fit(x_train_selected, y_train)
    y_head = classifier.predict(x_test_selected)
    score = accuracy_score(y_test, y_head)
    print("Fold No : {} - Feature Count :{} - Accuracy {} ".format(fold,featureCount,score))
    break
```
